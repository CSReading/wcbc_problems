## 5.7 

入力からすべてのGを取り除くチューリングマシンを作れ。


### 解答

複雑なので、tmファイルは省略。

- 文字列を1文字読み込みyに変更→末尾に読み込んだ文字を追加し、前に戻っていく
    - 初回については末尾+1つの空文字のあとに文字を追加する
        - 終点を明らかにするため
    - 読み込んだ文字列がGであればyに変更するが、末尾には追加しない
- 空白文字の手前までの文字列がすべてyになった場合、yをすべて空文字にする
- 最後に空白文字以降に残った文字列を左にシフトさせていく
    - `shiftinteger`と同等
