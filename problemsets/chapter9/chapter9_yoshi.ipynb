### 9-7 (2)

### 問題
- 入力：状態$Q=\{q_0, q_1, \dots \}$をもつNFA. $qr \notin Q$.
- 出力：状態$S=\{s_0, s_1, \dots \}$をもつ狭義NFA.$qr \notin Q$.

### アルゴリズム
1. $q_{(i)} \to q_{(j)}: \varepsilon$となる$(q_{(i)}, q_{(j)})$が存在するか確認し存在しなければ,
- $|Q| = n$, に対して, $S = \{s_0, \dots s_{n-1}\}$とする.
- $s_0$は記号を一切読まずにNFAが到達できる状態とする
- $\forall a \in \Sigma,  \delta(q_{(i)}, a) =q_{(j)} \Rightarrow \delta'(s_{i}, a) = s_{j}$

として(S, \delta')を出力.
存在する場合は2へ進む.

2. 
(a)$q_i \to q_j: \varepsilon$という遷移が存在する$(q_i,q_j) \in Q^2$を任意に固定する. $NextQ_j = \{q \in Q : \exists a \in \Sigma, \delta(q_j, a) = q \}$と定義する.
このとき, $q \in NextQ_j, \delta(q_j, a) = q$を満たす全ての$(q, a) \in Q \times \Sigma \cup \{ \varepsilon \}$に対して, $\delta$を下記のように更新する.

- $\delta(q_i, a) = q$
- $\delta(q_i, \varepsilon) = qr$

(b) $\delta(q, a) = q_j$となる$(q, a) \in Q \times \Sigma$ が存在しない場合, $q_j$をQから取り除く.

1へ戻る.
