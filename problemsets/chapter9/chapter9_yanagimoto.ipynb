## 9.4（柳本）
3の倍数（と空文字列）
#### 説明
- このDFAは三つの状態があり、各桁の数字によって状態が遷移する。
- 3の倍数(0, 3, 6, 9)で状態が遷移せず、3で割った余り1(1, 4, 7)で状態が一つ、余り2(2, 5, 8)で二つ状態が遷移する。
- 最終的に、状態0に戻った場合、入力を受理する。

したがって、これは桁の数字の合計が3の倍数であるかを判定する。
これは3の倍数の必要十分条件であるので、このDFAは3の倍数を判定する。

```python
tm_to_diagram(rf('q9_4.dfa'))
```

```python
for i in range(20):
    res = simulateDfa(rf('q9_4.dfa'), str(i))
    print(f'{i}:{res}')
```

```python
simulateDfa(rf('q9_4.dfa'), '')
```
