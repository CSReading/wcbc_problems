### 8.5

```python
g = Graph()
g.attr('node', style='filled', fillcolor='lightgrey')
sol = ['no', 'GAGA','TATA', 'AAAA']

for i in range(len(sol)):
    g.edge('root', str(i+1))

g.attr('node', style='filled', color='white', fillcolor='white', fontsize='10')
for i in range(len(sol)):
    g.edge(str(i+1), sol[i], color='white')

g
```

### 8.10
**問.**  マルチコア量子コンピュータは，標準チューリングマシンとチューリング等価か<br>
**答.** チューリング等価<br>
主張8.1 の証明と同様に、マルチコア量子コンピュータが計算できる問題は、シングルコア量子コンピュータでも計算できる。
シングルコア量子コンピュータで計算できる問題がマルチコア量子コンピュータで計算できることは自明であるので（一つのコアのみを使えばよい）、
この二つはチューリング等価。
シングルコア量子コンピュータは標準チューリングマシンとチューリング等価であるので(5.7章)、マルチコア量子コンピュータは標準チューリングマシンとチューリング等価。
